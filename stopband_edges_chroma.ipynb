## helping functions


```python
def find_edges(dataframe, threshold, qy_ucut = 18.5, dQsc = 0.0, test=False, target="delta_emit_y", bunched=True):

    qy_lcut = 18.5 if bunched else qy_ucut
    df_lower = dataframe[dataframe["Qy"] < qy_lcut]
    df_lower = df_lower.sort_values(by=['Qy'])
    df_upper = dataframe[dataframe["Qy"] > qy_ucut]
    df_upper = df_upper.sort_values(by=['Qy'])

    ql = np.linspace(df_lower["Qy"].min(),df_lower["Qy"].max(),100)
    qu = np.linspace(df_upper["Qy"].min(),df_upper["Qy"].max(),100)
    
    delta_lower = np.interp(x=ql, xp=df_lower["Qy"],fp=df_lower[target])
    delta_upper = np.interp(x=qu, xp=df_upper["Qy"],fp=df_upper[target])

    for q,d in zip(ql, delta_lower):
        lower_edge = q
        if d > threshold: #* q / (q - dQsc):
            break;

    for q,d in zip(qu, delta_upper):
        threshold_upper = threshold * q / (q - dQsc)
        if test:
            print(threshold_upper)
        upper_edge = q
        if d < threshold_upper:
            break;
            
    return lower_edge, upper_edge
```

## Threshold

```python
threshold = .005
```
